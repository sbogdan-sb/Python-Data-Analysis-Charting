```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
from matplotlib.ticker import PercentFormatter


# Hide warning messages in notebook
import warnings
warnings.filterwarnings("ignore")

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data_df = pd.read_csv(mouse_drug_data_to_load)
clinical_data_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merged_df = pd.merge(mouse_data_df, clinical_data_df, how = "right", on = "Mouse ID")

# Display the data table for preview
merged_df.head()
```

```python
#Clean the data

print(mouse_data_df["Mouse ID"].value_counts().head())
print("")
print(mouse_data_df[mouse_data_df["Mouse ID"] == "g989"]["Drug"])

```

```python
#Remove mouse g989 from data set since it was tested with 2 drugs
merged_df = merged_df[merged_df["Mouse ID"] != "g989"]
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_vol = merged_df.groupby(["Drug", "Timepoint"]).mean()
# del mean_tumor_vol["Metastatic Sites"]

# Convert to DataFrame
del mean_tumor_vol["Metastatic Sites"]
# Preview DataFrame
mean_tumor_vol.head(10)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
mean_tumor_sem = merged_df.groupby(["Drug", "Timepoint"]).sem()

# Convert to DataFrame
del mean_tumor_sem["Metastatic Sites"]
del mean_tumor_sem["Mouse ID"]

# Preview DataFrame
mean_tumor_sem.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
mean_tumor_vol_transposed = pd.pivot_table(mean_tumor_vol, index=["Timepoint", "Drug"]).unstack()
mean_tumor_vol_transposed.rename(columns={"Tumor Volume (mm3)":""}, inplace = True)

mean_tumor_vol_transposed.head(5)
```

```python
mean_tumor_sem_transposed = pd.pivot_table(mean_tumor_sem, index=["Timepoint", "Drug"]).unstack()
mean_tumor_sem_transposed.rename(columns={"Tumor Volume (mm3)":""}, inplace = True)

mean_tumor_sem_transposed.head(5)
```

```python
# Generate the Plot (with Error Bars)
# Save the Figure

count_chart = mean_tumor_vol_transposed.plot(kind = "line", yerr = mean_tumor_sem_transposed,
                                            grid = True, figsize = (15,8), markersize = 10)


plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")

#Clean up multiindex key so just drug name will be displayed on chart and configure legend
legend_list = mean_tumor_vol_transposed.keys().get_level_values(1)
plt.legend(legend_list, ncol= 2, loc = "upper left", prop={"size" : 15})

#Add markers
marker_list = ("o", "v", "s", "x", "P", "D", "H", "d", "*", ">")

for i, line in enumerate(count_chart.get_lines()):
    line.set_marker(marker_list[i])
    
plt.savefig("Tumor Size Change.png")

plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 

# Convert to DataFrame

# Preview DataFrame

mean_metastatic = merged_df.groupby(["Drug", "Timepoint"]).mean()
del mean_metastatic["Tumor Volume (mm3)"]
mean_metastatic.head(10)
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

# Convert to DataFrame

# Preview DataFrame

mean_metastatic_sem = merged_df.groupby(["Drug", "Timepoint"]).sem()
del mean_metastatic_sem["Tumor Volume (mm3)"]
del mean_metastatic_sem["Mouse ID"]
mean_metastatic_sem.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked

mean_metastatic_transposed = pd.pivot_table(mean_metastatic, index=["Timepoint", "Drug"]).unstack()
mean_metastatic_transposed.rename(columns = {"Metastatic Sites" : ""}, inplace = True)
mean_metastatic_transposed.head(5)
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
mean_metastatic_sem_transposed = pd.pivot_table(mean_metastatic_sem, index=["Timepoint", "Drug"]).unstack()
mean_metastatic_sem_transposed.rename(columns = {"Metastatic Sites" : ""}, inplace = True)
mean_metastatic_sem_transposed.head(5)
```

```python
# Generate the Plot (with Error Bars)

# Save the Figure

# Show the Figure
metastatic_chart = mean_metastatic_transposed.plot(kind = "line", yerr = mean_metastatic_sem_transposed,
                                            grid = True, figsize = (15,8), markersize = 10)


plt.xlabel("Time (Days)")
plt.ylabel("Met. Sites")
plt.title("Metastatic Spread During Treatment")

#Clean up multiindex key so just drug name will be displayed on chart and configure legend
legend_list = mean_metastatic_transposed.keys().get_level_values(1)
plt.legend(legend_list, ncol= 2, loc = "upper left", prop={"size" : 15})

#Add markers
marker_list = ("o", "v", "s", "x", "P", "D", "H", "d", "*", ">")

for i, line in enumerate(metastatic_chart.get_lines()):
    line.set_marker(marker_list[i])
    
plt.savefig("Metastatic Spread.png")

plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

# Convert to DataFrame

# Preview DataFrame

mouse_lifespan = merged_df.groupby(["Drug", "Timepoint"]).count()
del mouse_lifespan["Tumor Volume (mm3)"]
del mouse_lifespan["Metastatic Sites"]

mouse_lifespan.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview the Data Frame

mouse_lifespan_transposed = pd.pivot_table(mouse_lifespan, index=["Timepoint", "Drug"]).unstack()
mouse_lifespan_transposed.rename(columns = {"Mouse ID": ""}, inplace = True)
mouse_lifespan_transposed.head(5)
```

```python
# Generate the Plot (Accounting for percentages)

# Save the Figure

# Show the Figure
lifespan_chart = mouse_lifespan_transposed.plot(kind = "line", grid = True, figsize = (15,8), markersize = 10)


plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.title("Survival During Treatment")

#Clean up multiindex key so just drug name will be displayed on chart and configure legend
legend_list = mouse_lifespan_transposed.keys().get_level_values(1)    
plt.legend(legend_list, ncol= 2, loc = "lower left", prop={"size" : 15})

#Add markers
marker_list = ("o", "v", "s", "x", "P", "D", "H", "d", "*", ">")

for i, line in enumerate(lifespan_chart.get_lines()):
    line.set_marker(marker_list[i])

#Plot as percentage
plt.gca().yaxis.set_major_formatter(PercentFormatter(25))

plt.savefig("Survival Rate.png")

plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

# Display the data to confirm
percent_change = ((mean_tumor_vol_transposed.iloc[9] - mean_tumor_vol_transposed.iloc[0]) /
                  mean_tumor_vol_transposed.iloc[0]) * 100
percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple

# Splice the data between passing and failing drugs

# Orient widths. Add labels, tick marks, etc. 

# Use functions to label the percentages of changes

# Call functions to implement the function calls

# Save the Figure

# Show the Figure

# fig.show()

percent_change_tuple = tuple(percent_change.values)

#Clean up multiindex key so just drug name will be displayed on chart and configure legend
legend_list = tuple(percent_change.keys().get_level_values(1))

fig, ax = plt.subplots()
tumor_bar_chart = ax.bar(legend_list, percent_change_tuple, color='r', width=1, alpha=0.5, align="center")

plt.title("Tumor Change Over 45 Day Treatment")
plt.xlabel("Drug")
plt.ylabel("% Tumor Volume Change")

plt.rcParams["figure.figsize"] = [20,4]

plt.grid(True)

#Function to print tumor size change values on corresponding bar on chart and set color based on drug pass or fail
def label_bars(bar_plot):

    for bar in bar_plot:
        height = bar.get_height()
        if height < 0:
            y_position = -8
            bar.set_color("g")
        else: 
            y_position = 2
        
        
        ax.text(bar.get_x() + bar.get_width()/2, y_position, 
               "%0.2f" % height, ha = "center", va = "bottom", color = "w", fontsize = 15)
        
#Call label_bar function to label bars on both charts  
label_bars(tumor_bar_chart)

plt.savefig("Tumor Size Change (Bar).png")
plt.show()
```
